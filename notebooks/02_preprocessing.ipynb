```python
import pandas as pd
import re
```

```python
df = pd.read_parquet(r'C:\Projects\Recomender System\data\processed\ratings_movies.parquet')

df.columns
```

```python
df.head()
```

```python
df['clean_genres'] = df['genres'].str.replace('|', ' ', regex=False)
```

```python
def clean_title(title):
    return re.sub(r'\(\d{4}\)', '', title).strip()

df['clean_title'] = df['title'].apply(clean_title)
```

```python
df['text'] = df['clean_title'] + ' ' + df['clean_genres']
```

```python
df_movies_unique = df[['movieId', 'clean_title', 'clean_genres', 'text']].drop_duplicates('movieId')
print(df_movies_unique.shape)
df_movies_unique.head()
```

```python
df_movies_unique.to_parquet(r'C:\Projects\Recomender System\data\processed\movies_for_content_based.parquet', index=False)
```
